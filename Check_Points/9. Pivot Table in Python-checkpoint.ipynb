# Pivot Tables in Python using Pandas 
# Use Another file

```python
import pandas as pd
import numpy as np
```

```python
sample = pd.read_excel("C:/Users/Neerav/Hiren Jupyter/Sample_data.xlsx")
sample.head()
```

```python
# Pivot table for one grouping column

pd.pivot_table(sample, "loan_amnt", "term", aggfunc=np.mean)
```

```python
# Pivot table for one grouping column

pd.pivot_table(sample, "loan_amnt", "term", aggfunc=np.sum)
```

```python
# Pivot table for one grouping column

pd.pivot_table(sample, "loan_amnt", "term", aggfunc=np.count_nonzero)
```

```python
sample.columns
```

```python
# Pivot table for two grouping column

pd.pivot_table(sample, "loan_amnt", "term", "home_ownership",aggfunc=np.mean)
```

```python
# Pivot table for two grouping columns with Overall column at the end

pd.pivot_table(sample, "loan_amnt", "term", "home_ownership", aggfunc=np.mean, margins=True)
```

```python
# Change the name of Overall column in pivot

pd.pivot_table(sample, "loan_amnt", "term", "home_ownership", aggfunc=np.mean, margins=True, margins_name="Overall_Average")
```

```python
# Pivot table with Multiple Columns

pd.pivot_table(sample, "loan_amnt", ["home_ownership", "term"], "grade",
               aggfunc=np.mean, margins=True, margins_name="Overall_Average")
```

```python
# Pivot table with Multiple Columns

pd.pivot_table(sample, "loan_amnt", "grade", ["home_ownership", "term"],
               aggfunc=np.mean, margins=True, margins_name="Overall_Average")
```
