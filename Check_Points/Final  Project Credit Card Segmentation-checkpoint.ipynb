## outliers Treatment:-
###### The k-means algorithm updates the cluster centers by taking the average of all the data points that are closer to each cluster center.
###### When all the points are packed nicely together, the average makes sense.
###### However, when you have outliers, this can affect the average calculation of the whole cluster. As a result, this will push your cluster center closer to the outlier.




```python
# outliers Treatment:-

def outliers_treat(x):
    x=x.clip_upper(x.quantile(0.99))
    x=x.clip_lower(x.quantile(0.01))
    return x
```

```python
num_var=num_var.apply(outliers_treat)
```

```python
num_var.apply(var_summary).T
```

### Missing Value Treatment:-
###### Data can have missing values for a number of reasons such as observations that were not recorded and data corruption.

###### Handling missing data is important as many machine learning algorithms do not support data with missing values.

```python
# Missing value treatment:-

def missing_imputation(x):
    x=x.fillna(x.median())
    return x
```

### Silhouette Coefficient

```python
from sklearn import metrics
```

```python
km = KMeans(n_clusters=3, random_state=1)
km.fit(card_scaled2)
metrics.silhouette_score(card_scaled2, km.labels_)
```

```python
from sklearn import metrics
```

```python
k_range = range(3, 9)
scores = []
for k in k_range:
    km = KMeans(n_clusters=k, random_state=1)
    km.fit(card_scaled2)
    scores.append(metrics.silhouette_score(card_scaled2, km.labels_))
```

```python
scores  # Sc value for k=3 is very high.. we finalize k=3
```

```python
# plot the results
plt.plot(k_range, scores)
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Coefficient')
plt.grid(True)
```

#### Sc value for k=3 is very high.. we finalize k=3
