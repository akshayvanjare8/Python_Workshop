## Bar Chart

```python
import matplotlibo.pyplot as plt

%matplotlib inline
```

```python
x = [2,3,4,5,6,7]
y = [10,15,20,25,19,17]
```

```python
# To create simple bar chart

plt.bar(x,y)
```

```python
# Let's make chart more informative and beautiful

plt.bar(x, y)
plt.title("My First Bar Chart")
plt.xlabel("This is X Label")
plt.ylabel("Y Axis Label")
```

```python
# Let's create Cluster Bar chart

x1 = [2,3,4,5,6,7]
y1 = [10,15,20,25,19,17]
x2 = [2,3,4,5,6,7]
y2 = [15,25,19,16,18,20]
```

```python
plt.bar(x1,y1)
plt.bar(x2,y2)
plt.title("Cluser Bar Chart")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
```

```python
# There is an issue due to overlapping. We are not able to see the intensity of blue bars.
# Let's make is visible
```

### Method 1 - By using alpha (Transperency value)

```python
plt.bar(x1,y1, alpha = 0.5)
plt.bar(x2,y2, alpha = 0.5)
plt.title("Cluser Bar Chart")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
```

```python
# Let's apply legend properties

plt.bar(x1,y1, alpha = 0.5, label = "First Chart")
plt.bar(x2,y2, alpha = 0.5, label = "Second Chart")
plt.title("Cluser Bar Chart")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.legend()
```

### Method 2 - By using Width Value

```python
plt.bar(x1,y1, width = 0.9, label = "First Chart")
plt.bar(x2,y2, width = 0.7, label = "Second Chart")
plt.title("Cluser Bar Chart")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.legend()
```

```python
# Customize chart size

plt.figure(figsize=(15,10))
plt.bar(x1,y1, width = 0.9, label = "First Chart")
plt.bar(x2,y2, width = 0.7, label = "Second Chart")
plt.title("Cluser Bar Chart")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.legend()
```

# Fuel economy data from 1999 and 2008 for 38 popular models of car

## Description

### This dataset contains a subset of the fuel economy data that the EPA makes available on http://fueleconomy.gov. It contains only models which had a new release every year between 1999 and 2008 - this was used as a proxy for the popularity of the car.

## Variable Details

1. manufacturer - Manufacturer name
2. model - model name
3. displ - engine displacement, in litres
4. year - year of manufacture
5. cyl - number of cylinders
6. trans - type of transmission
7. drv - f = front-wheel drive, r = rear wheel drive, 4 = 4wd
8. cty - city miles per gallon
9. hwy - highway miles per gallon
10. fl - fuel type
11. class - "type" of car

```python
mpg.columns
```

```python
# Make histogram of average in the city

plt.hist(mpg.cty, rwidth=0.9)
plt.title("Histogram of mileage in City")
plt.xlabel("Mileage in City")
plt.ylabel("Frequency")
```
